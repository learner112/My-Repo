```python
import time
import matplotlib.pyplot as plt

def linear_time_example(lst):
    for item in lst:
        print(item)

def measure_time(func, input_sizes):
    times = []
    
    for size in input_sizes:
        lst = list(range(size))
        
        start_time = time.time()
        func(lst)
        end_time = time.time()
        
        times.append(end_time - start_time)
    
    return times

input_sizes = [1000, 2000, 3000, 4000, 5000]
times = measure_time(linear_time_example, input_sizes)

plt.plot(input_sizes, times)
plt.xlabel('Input Size')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity Analysis')
plt.show()
```
